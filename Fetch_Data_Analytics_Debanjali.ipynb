## Import libraries and read jsons as pandas dataframes

```python
import pandas as pd
import json
import numpy as np
import re
from pandas import json_normalize

import warnings
warnings.filterwarnings("ignore")
```

```python
users = pd.read_json('users.json', lines= True)
brands = pd.read_json('brands.json', lines= True)
receipts = pd.read_json('receipts.json', lines= True)
```

```python
# Renaming all id columns for each dataframe to avoid errors and confusion
users.rename(columns={'_id':'user_id'}, inplace=True)
brands.rename(columns={'_id':'brand_id'}, inplace=True)
receipts.rename(columns={'_id':'receipt_id'}, inplace=True)
```

## Users data pre-processing

```python
users.head()
```

```python
# Extracting user_id, createdDate, and lastLogin in clean format
users['user_id'] = users['user_id'].apply(lambda x: x.get('$oid'))
users['createdDate'] = users['createdDate'].apply(lambda x: x.get('$date') if str(x) != str(np.nan) else x)
users['lastLogin'] = users['lastLogin'].apply(lambda x: x.get('$date') if str(x) != str(np.nan) else x)
     
users.head()
```

```python
# Converting createdDate and lastLogin columns to datetime type from milliseconds
users['createdDate'] = pd.to_datetime(users['createdDate']//1000, unit='s') # convert to datetime
users['lastLogin'] = pd.to_datetime(users['lastLogin']//1000, unit='s')

users.head()
```

## Brands data pre-processing

```python
brands.head()
```

```python
# Converting cpg from nested dictionary into a dataframe
cpg_df = pd.DataFrame([x for x in brands['cpg']])

cpg_df
```

```python
# Renaming columns
cpg_df.rename(columns={'$id':'cpg_id', '$ref':'cpg_ref'}, inplace=True)

cpg_df.head()
```

```python
# Extracting cpg_id from $oid
cpg_df['cpg_id'] = cpg_df['cpg_id'].apply(lambda x: x.get('$oid'))

cpg_df.head()
```

```python
# Extracting brand_id in clean format
brands['brand_id'] = brands['brand_id'].apply(lambda x: x.get('$oid'))

brands.head()
```

```python
# Merging into a single dataframe and dropping the original cpg column
brands = pd.merge(brands, cpg_df, left_index=True, right_index=True)
brands.drop(columns='cpg', inplace=True)

brands.head()
```

```python
# Checking if all cpg_ref values are same
brands['cpg_ref'].unique()
```

```python
# Combining cpg_id and cpg_ref into one column called 'cpg' so it aligns with the ER diagram 
brands['cpg'] = brands['cpg_id'].astype(str) + ":" + brands['cpg_ref']
brands= brands.drop(['cpg_id', 'cpg_ref'], axis=1)

brands
```

## Receipts data pre-processing

```python
receipts.head()
```

```python
# Extracting receipt_id in clean format
receipts['receipt_id'] = receipts['receipt_id'].apply(lambda x: x.get('$oid'))

receipts.head()
```

```python
# Extracting all the date columns in Receipts data in datetime format
for i in receipts.columns:
    if re.findall(r'date|Date', i):
        receipts[i] = receipts[i].apply(lambda x: x.get('$date') if str(x) != str(np.nan) else x)
        receipts[i] = pd.to_datetime(receipts[i]//1000, unit='s')

receipts.head()
```

### Creating the ITEM entity table for the database

```python
# Checking content of 'rewardsReceiptItemList' column for first record
receipts['rewardsReceiptItemList'][0]
```

```python
# Checking content of 'rewardsReceiptItemList' column for second record
receipts['rewardsReceiptItemList'][1]
```

```python
# Creating a new dataframe called 'items' and parsing all items in each receipt from the 'receipts' dataframe.
items = pd.DataFrame()

for i in receipts.index:
    try:
        for json in receipts['rewardsReceiptItemList'][i]:
            try:
                temp = pd.DataFrame([json])
                items = pd.concat([items, temp], ignore_index=True)
            except:
                # print("failed to read:", i) ## Many values here could not be read because they are NaN values - DATA QUALITY ISSUE
                pass
    except:
        # print("failed to read:", i) ## Many values here could not be read because they are NaN values - DATA QUALITY ISSUE
        pass
items
```

```python
# Creating the ITEM entity of the ER diagram
ITEM = items.groupby(['barcode','description', ])['itemPrice'].apply(', '.join).reset_index()

ITEM= ITEM.drop('itemPrice', axis=1) # dropping the itemPrice column since itemPrice is different across receipts

ITEM
```

```python
# Creating a new dataframe called 'brand_item' to be merged with ITEM entity table to specify brand_id for each unique item barcode
brand_item = brands[['barcode','brand_id']].copy()

brand_item
```

#### Checking if datatypes across the two dataframes are consistent for the merge

```python
ITEM.dtypes
```

```python
brand_item.dtypes
```

```python
brand_item['barcode'] = brand_item['barcode'].apply(str)

brand_item.dtypes
```

```python
# Merging the two dataframes to finally create the complete ITEM entity table for the database
ITEM = pd.merge(ITEM, brand_item, on='barcode', how='left')

ITEM
```

### Creating the REWARD_EVENT entity table for the database

```python
receipts.head()
```

```python
# Creating a new dataframe called 'rewards_events' that contains all reward details for each item from each receipt
rewards_events= pd.DataFrame()

for i in receipts.index:
    # print(receipts['rewardsReceiptItemList'][i])
    try:
        for json in receipts['rewardsReceiptItemList'][i]:
            # print(json)
            try:
                # rewards_events['receipt_id'][i]= receipts['receipt_id'][i]
                temp = pd.DataFrame([json])
                temp['receipt_id']= receipts['receipt_id'][i]
                rewards_events = pd.concat([rewards_events, temp], ignore_index=True)
            except:
                # print("failed to read:", i)
                pass
    except:
        # print("failed to read:", i)
        pass

rewards_events
```

```python
rewards_events.columns
```

```python
# Moving receipt_id column to extreme left

column_to_move = rewards_events.pop("receipt_id")
rewards_events.insert(0, "receipt_id", column_to_move)

rewards_events
```

```python
# Creating the REWARD_EVENT entity table in the ER diagram containing all attributes from the ITEM side

REWARD_EVENT = rewards_events.drop('description', axis=1)

REWARD_EVENT
```

```python
# Creating a new dataframe called 'receipt_details' containing all reward event based attributes of each receipt
# (to be merged with REWARD_EVENT entity table
receipt_details= receipts[['receipt_id', 'bonusPointsEarnedReason', 'createDate', 'modifyDate']].copy()

receipt_details
```

```python
# Merging the two dataframes
REWARD_EVENT = pd.merge(REWARD_EVENT, receipt_details, on='receipt_id', how='left')

REWARD_EVENT
```

```python
REWARD_EVENT.columns
```

```python
# Dropping brandCode attribute to finally have the REWARD_EVENT entity table ready
REWARD_EVENT= REWARD_EVENT.drop('brandCode', axis=1)

REWARD_EVENT.columns
```

```python
# Dropping the attriutes shifted to the REWARD_EVENT table to create the RECEIPT entity table for the database
RECEIPT = receipts.drop(['bonusPointsEarnedReason','createDate', 'modifyDate', 'rewardsReceiptItemList'], axis=1)

RECEIPT
```

# Second: Write a query that directly answers a predetermined question from a business stakeholder

```python
import sqlite3

# Creating an in-memory SQLite database
connection = sqlite3.connect(':memory:')

# Loading the dataframe as tables in the database
USER.to_sql('user', connection, index=False)
BRAND.to_sql('brand', connection, index=False)
CATEGORY.to_sql('category', connection, index=False)
RECEIPT.to_sql('receipt', connection, index=False)
REWARD_EVENT.to_sql('reward_event', connection, index=False)
ITEM.to_sql('item', connection, index=False)
```

### 1) What are the top 5 brands by receipts scanned for most recent month?

```python
query1 = """
SELECT b.name, COUNT(r.receipt_id) AS receipts_scanned
FROM brand b
JOIN item i ON b.brand_id = i.brand_id
JOIN reward_event rv ON i.barcode = rv.barcode
JOIN receipt r ON rv.receipt_id = r.receipt_id
WHERE r.dateScanned >= '2021-01-01 00:00:00'
GROUP BY b.name
ORDER BY receipts_scanned DESC
LIMIT 5;"""

result1 = pd.read_sql_query(query1, connection)

result1
```

### 5) Which brand has the most spend among users who were created within the past 6 months?

```python
query5 = """
SELECT b.name, SUM(r.totalSpent) AS total_spent, u.createdDate
FROM brand b
JOIN item i ON b.brand_id = i.brand_id
JOIN reward_event rv ON i.barcode = rv.barcode
JOIN receipt r ON rv.receipt_id = r.receipt_id
JOIN user u ON r.userId = u.user_id
WHERE u.createdDate >= '2020-08-01 00:00:00'
GROUP BY b.name
ORDER BY total_spent DESC
LIMIT 1; """

result5 = pd.read_sql_query(query5, connection)

result5
```

### 6) Which brand has the most transactions among users who were created within the past 6 months?

```python
query6 = """
SELECT b.name, COUNT(*) AS transaction_count, u.createdDate
FROM brand b
JOIN item i ON b.brand_id = i.brand_id
JOIN reward_event rv ON i.barcode = rv.barcode
JOIN receipt r ON rv.receipt_id = r.receipt_id
JOIN user u ON r.userId = u.user_id
WHERE u.createdDate >= '2020-08-01 00:00:00'
GROUP BY b.name
ORDER BY transaction_count DESC
LIMIT 1;"""

result6 = pd.read_sql_query(query6, connection)

result6
```
